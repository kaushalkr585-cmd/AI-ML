```python
df["gender"].str.capitalize()
df["gender"].str.upper()
df["name"].str.split(" ")
df["country"].str.contains("india", case=False)
```

```python
df
```

```python
df2 = df.copy()

# Create tax column
df2["tax"] = df2["income"].apply(lambda x: "20%" if x >= 50000 else "10%")

# Map gender values
gender_map = {
    "Male": "M",
    "Female": "F",
    "Unknown": "U"
}

df2["gender"] = df2["gender"].map(gender_map)

# Create new_income column
df2 = df2.assign(new_income=df2["income"] * 1.1)

# Replace USA with US
df2["country"] = df2["country"].replace("USA", "US")

# Rename columns
df2.columns = [
    "Id",
    "Name",
    "Age",
    "Country",
    "Gender",
    "Income",
    "Tax",
    "New_Income"
]

# Rename column
df2 = df2.rename(columns={"Income": "Salary"})

# Rename index
df2 = df2.rename(index={1: "First"})

# Sort by Salary
df2 = df2.sort_values("Salary")

# Fill missing values
df2 = df2.fillna(50)

# Multiple column sorting
sorted_df = df2.sort_values(["Salary", "Age"])

# Sort index
sorted_df = sorted_df.sort_index()

# Reset index
sorted_df = sorted_df.reset_index(drop=True)

print(sorted_df)

# Ranking 

sorted_df["Ranking"]=sorted_df["Income"].rank(ascending = False , method="dense")
```

```python
# Grouping and Aggregation of Data 


df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()
df.groupby("gender")["income"].mean()
df.groupby("country")["income"].agg(["mean","min","max"])
df.groupby("country")["income"].agg(avg_sal="mean",min_sal="min",max_sal="max")
df.groupby("country").agg({
    "income" : "max",
    "age": "mean",
})
```

```python
# Melt & Pivot 
 
df = pd.DataFrame({
    "country": ["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22]
})

melted_df=df.melt(
    id_vars=["country","year"],
    value_vars=["sales","profit"],
    var_name="metrics",
    value_name="value"
)

melted_df.pivot(
    index=["country","year"],
    columns="metrics",
    values="value"
)
```

```python
df_customers = pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})

def_orders = pd.DataFrame({
    "order_id" : [101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount" : [250,120,300,100]
})

pd.merge(df_customers,def_orders, on="customer_id") #inner join 
pd.merge(df_customers,def_orders, on="customer_id",how="left")
pd.merge(df_customers,def_orders, on="customer_id",how="right")
pd.merge(df_customers,def_orders, on="customer_id",how="outer")
```
