```python
import json 
```

```python
#load the data 
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data=load_data("store_data.json")
print(data,type(data))
```

```python
#clean & structure the data 
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data=[]
    unique_users= set() #tracks unique data 
    for user in data:
        #clean ratings
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        print(raw_rating)

        #Handle Missing vals 

        raw_age = user.get("age")
        if(raw_age == None):
            user["age"]=None

        #Deduplication 
        if (user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
            
```

```python
data = clean_data(data)
```

```python
# get insights from data

def get_insights(data):

    # avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating / len(data)}")

    # percentage of users with poor ratings
    poor_ratings = 0

    for user in data:
        if float(user["rating"]) < 3:
            poor_ratings += 1

    print(f"% of users with poor rating = {(poor_ratings / len(data)) * 100}%")
```

```python
get_insights(data)
```

```python
def get_recommendation(data):
    recommendations = []

    for user in data:
        curr_recom ={}
        curr_recom =["name"] = user["name"]
        if(float(user["rating"]) >= 4):
            curr_recom["brand"] = "Apple"
        else:
            cur_recom["brand"] = "Samsung"
        recommendations.append(curr_recom)
    return recommendations
```

```python
get_recommendation(data)
```
