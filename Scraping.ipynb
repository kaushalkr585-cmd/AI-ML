```python
import requests 
from bs4 import BeautifulSoup
import lxml
```

```python
#Fetch data & store it 

page_count = 1

while True:
    url = f"https://quotes.toscrape.com/page/{page_count}/"

    res = requests.get(url)
    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore...")
        break

    with open(
        f"scraped_data/quotes{page_count}.html",
        "w",
        encoding="utf-8"
    ) as f:
        f.write(res.text)

    print(f"downloaded data from page {page_count}")
    page_count += 1
```

```python
all_quotes = soup.select("div.quote")

life_quotes = []

for q in all_quotes:
    all_tags = []

    for tag in q.select(".tags .tag"):
        all_tags.append(tag.get_text())

    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()

        print(text)
        life_quotes.append([text, author])

print(life_quotes)
```

```python
from bs4 import BeautifulSoup

life_quotes = []

for page in range(1, 11):
    with open(f"scraped_data/quotes{page}.html", encoding="utf-8") as f:
        soup = BeautifulSoup(f.read(), "lxml")

    for q in soup.select("div.quote"):
        tags = [tag.get_text() for tag in q.select(".tag")]

        if "life" in tags:
            text = q.select_one(".text").get_text()
            author = q.select_one(".author").get_text()

            life_quotes.append([text, author])

print(f"Total life quotes: {len(life_quotes)}")
print(life_quotes)
```
